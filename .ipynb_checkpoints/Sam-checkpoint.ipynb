```python
state = input('Enter State: ex. NewYork ')
state = pd.read_csv(state)
state
```

```python
#filter by campus housing
from IPython.display import display

campus_housing = input('Do you want campus housing? Yes or no or press enter if you do not have a preference')
campus_housing = campus_housing.capitalize()
print(campus_housing)
if campus_housing == 'Yes':
    state6 = state[state['Campus housing'] == 'Yes']
    display(state6)

elif campus_housing == 'No':
    state6 = state[state['Campus housing'] == 'No']
    display(state6)
else: 
    state6 = state
    display(state6)
    
```

```python
from IPython.display import display


college_types = list(state['Type'].value_counts().index)
i = 1
for types in college_types:
    print(i,types)
    i = i + 1
choice = int(input ('What type of university are you interested in? Enter the correspondong number (1 - 6). '))
choice_value = college_types[choice-1]
state2 = state[state['Type']==choice_value]
display(state2)
```

```python
#filter by campus setting

surrounding_area = list(state['Campus setting'].value_counts().index)
i = 1
for types in surrounding_area:
    print(i,types)
    i = i + 1
area = int(input ('What setting would you like the campus to be in? Please choose from the menu by entering the corresponding number.'))
setting_preference = surrounding_area[area-1]
state3 = state[state['Campus setting']==setting_preference]
display(state3)
```

```python
state['size'] = np.nan
state['size'][state['Student population']>15000] = 'large'
state['size'][state['Student population']<=15000] = 'medium'
state['size'][state['Student population'] <=5000] = 'small'
state

college_sizes = list(state['size'].value_counts().index)
i = 1
for types in college_sizes:
    print(i,types)
    i = i + 1
size_preference = input("How big do you want your university to be? Above are your options. ")
state4=state[state['size']==size_preference]
state4

```
